This notebook contains code to load and preprocess the data. The data is then passed to different classification models for training in the same notebook.

Firstly, let's import all the required libraries and modules.

```python
from src.Preprocessing import load_data, preprocess_data
from src.train_models import train_logistic_regression, train_random_forest, train_gradient_boosting, train_decision_tree, save_model
from src.logreg_gd import grid_search_logistic_regression
```

```python
file_path = 'Dataset/creditcard.csv'  
data = load_data(file_path)
(X_train, y_train), (X_test, y_test), (X_train_over, y_train_over), (X_train_smote, y_train_smote), (X_train_under, y_train_under) = preprocess_data(data)
```

Now, we will train different classification models.

For starters, we have created different sets of data for ourselves.

X_train,y_train --> represent the original data without any smapling applied on it.<br>
X_train_under,y_train_under --> represent the Undersampled training data using the RandomUnderSampler.<br>
X_train_over,y_train_over --> represent the Oversampled training data using the RandomOverSampler.<br>
X_train_smote,y_train_smote --> represent the sampled data using SMOTE.<br>

Now we have differently sampled datasets, we will be training our models on all the datasets to see which performs the best.<br>

Also, as we are using GridSearchCV we will be training each model multiple times, each time using a different scoring(f1, precision, recall, roc_auc) to see if the best parameters selected by GridSearchCV change and to see which scoring performs better. Make sure to change the scoring rate accordingly in the training function in train_models.py file before training the model. By default it is set to f1_scoring.<br>

We will start with the Sklearns LogisticRegression model with original data and different GridSearch scorings.<br> 

```python
model1 = train_logistic_regression(X_train, y_train)
save_model(model1, 'logistic_regression_f1', save=True)

model2 = train_logistic_regression(X_train, y_train)
save_model(model2, 'logistic_regression_precision', save=True)

model3 = train_logistic_regression(X_train, y_train)
save_model(model3, 'logistic_regression_recall', save=True)

model4 = train_logistic_regression(X_train, y_train)
save_model(model4, 'logistic_regression_roc', save=True)
```
